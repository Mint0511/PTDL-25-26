1.2.1. Bài làm mẫu

Chuẩn bị dữ liệu cho trực quan hoá dữ liệu

```python
import pandas as pd
houseprices_data = pd.read_csv("data/HousingPricesData.csv")
houseprices_data = houseprices_data[['Zip', 'Price', 'Area', 'Room']]
# Create a PriceperSqm variable based on the Price and Area variables:
houseprices_data['PriceperSqm'] = houseprices_data['Price']
houseprices_data['Area']
```

Trực quan hoá dữ liệu với thư viện Matpotlib

```python
import matplotlib.pyplot as plt
houseprices_sorted = houseprices_data.sort_values('Price', ascending =
False)
houseprices_sorted.head()
# case 1: basic
plt.figure(figsize= (12,6))
x = houseprices_sorted['Zip'][0:10]
plt.xticks(fontsize=10)
y = houseprices_sorted['Price'][0:10]
plt.yticks(fontsize=10)
plt.bar(x,y)
plt.savefig('/content/drive/MyDrive/PTDL 25-26/lab01-Xử lý trực quan hoá dữ liệu/img/Ví dụ mẫu/case1_matplotlib.png')
plt.show()
# case 2: advanced 1
plt.figure(figsize= (12,6))
plt.bar(x,y)
plt.title('Top 10 Areas with the highest house prices', fontsize=15)
plt.xlabel('Zip code', fontsize = 12)
plt.xticks(fontsize=10)
plt.ylabel('House prices in millions', fontsize=12)
plt.yticks(fontsize=10)
plt.savefig('/content/drive/MyDrive/PTDL 25-26/lab01-Xử lý trực quan hoá dữ liệu/img/Ví dụ mẫu/case2_matplotlib.png')
plt.show()
# case 3: advanced 2
fig, ax = plt.subplots(1, 2, figsize=(24,8))
x = houseprices_sorted['Zip'][0:10]
y = houseprices_sorted['Price'][0:10]
y1 = houseprices_sorted['PriceperSqm'][0:10]
ax[0].bar(x,y)
ax[0].tick_params(axis='x', labelsize=17)
ax[0].set_ylabel('House prices in millions', fontsize=25)
ax[0].tick_params(axis='y', labelsize=20)
ax[0].set_title('Top 10 Areas with the highest house prices',
fontsize=25)
ax[1].bar(x,y1)
ax[1].tick_params(axis='x', labelsize=17)
ax[1].set_ylabel('House prices per sqm', fontsize=25)
ax[1].tick_params(axis='y', labelsize=20)
ax[1].set_title('Top 10 Areas with the highest house prices per sqm',
fontsize=25)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/PTDL 25-26/lab01-Xử lý trực quan hoá dữ liệu/img/Ví dụ mẫu/case3_matplotlib.png')
plt.show()
```

**Nhận xét về Biểu đồ Giá Nhà và Giá trên mỗi mét vuông tại 10 khu vực đắt nhất**

Ba biểu đồ trên cung cấp cái nhìn sâu sắc về thị trường bất động sản tại các khu vực có giá nhà cao nhất, tập trung vào giá trị tuyệt đối của căn nhà và giá trị tương đối trên mỗi mét vuông.

Biểu đồ đầu tiên và thứ hai cho thấy Top 10 Khu vực có Giá nhà cao nhất dựa trên mã Zip. Rõ ràng, đây là những nơi tập trung các bất động sản có giá trị đầu tư lớn, thu hút sự chú ý của những người tìm kiếm các tài sản đắt đỏ. Các cột biểu thị trực tiếp mức giá triệu đô, phản ánh phân khúc cao cấp của thị trường.

Tuy nhiên, khi xem xét biểu đồ thứ ba, nơi so sánh Giá nhà (tuyệt đối) và Giá trên mỗi mét vuông (tương đối) tại cùng 10 khu vực này. Biểu đồ này hé lộ một khía cạnh quan trọng: một căn nhà có giá trị cao nhất không nhất thiết là căn nhà có giá trên mỗi mét vuông cao nhất. Điều này cho thấy sự đa dạng trong yếu tố định giá bất động sản, không chỉ dựa vào quy mô (diện tích) mà còn phụ thuộc vào các yếu tố khác như vị trí cụ thể trong khu vực Zip, tiện ích, hoặc đặc điểm riêng của căn nhà.

Sự chênh lệch giữa thứ hạng giá tuyệt đối và giá trên mỗi mét vuông tại một số mã Zip nhấn mạnh tầm quan trọng của việc xem xét cả hai chỉ số khi đánh giá giá trị thực của bất động sản. Đối với nhà đầu tư hoặc người mua, biểu đồ này là lời nhắc nhở rằng giá niêm yết chỉ là một phần của bức tranh; việc phân tích giá trên mỗi mét vuông sẽ cung cấp cái nhìn chính xác hơn về mật độ giá trị và tiềm năng sinh lời dựa trên diện tích sử dụng.

Tóm lại, các biểu đồ này không chỉ chỉ ra những khu vực "hot" nhất về giá trị tài sản, mà còn thể hiện sự khác biệt trong cấu trúc định giá, khuyến khích phân tích sâu hơn thay vì chỉ dựa vào giá trị bề mặt.

Trực quan hóa dữ liệu với thư viện Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
# case 1: basic
plt.figure(figsize= (12,6))
data = houseprices_sorted[0:10]
sns.barplot(data= data, x= 'Zip',y = 'Price')
plt.savefig('/content/drive/MyDrive/PTDL 25-26/lab01-Xử lý trực quan hoá dữ liệu/img/Ví dụ mẫu/case1_seaborn.png')
# case 2: advanced 1
plt.figure(figsize= (12,6))
data = houseprices_sorted[0:10]
ax = sns.barplot(data= data, x= 'Zip',y = 'Price')
ax.set_xlabel('Zip code',fontsize = 15)
ax.set_ylabel('House prices in millions', fontsize = 15)
ax.set_title('Top 10 Areas with the highest house prices', fontsize= 20)
plt.savefig('/content/drive/MyDrive/PTDL 25-26/lab01-Xử lý trực quan hoá dữ liệu/img/Ví dụ mẫu/case2_seaborn.png')
# case 3: view multiple perspectives at once
fig, ax = plt.subplots(1, 2,figsize=(20,6))
data = houseprices_sorted[0:10]
# sns.set(font_scale = 3)
ax1 = sns.barplot(data= data, x= 'Zip',y = 'Price', ax = ax[0])
ax1.set_xlabel('Zip code', fontsize = 14)
ax1.set_ylabel('House prices in millions', fontsize = 14)
ax1.set_title('Top 10 Areas with the highest house prices', fontsize = 16)
ax2 = sns.barplot(data= data, x= 'Zip',y = 'PriceperSqm',
ax=ax[1])
ax2.set_xlabel('Zip code', fontsize = 14)
ax2.set_ylabel('House prices per sqm', fontsize = 14)
ax2.set_title('Top 10 Areas with the highest price per sqm', fontsize = 16)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/PTDL 25-26/lab01-Xử lý trực quan hoá dữ liệu/img/Ví dụ mẫu/case3_seaborn.png')
plt.show()
```

**Nhận xét về Biểu đồ Giá Nhà và Giá trên mỗi mét vuông tại 10 khu vực đắt nhất**

Ba biểu đồ này tiếp tục làm sáng tỏ bức tranh về thị trường bất động sản tại những khu vực có giá nhà cao nhất.

Biểu đồ đầu tiên và thứ hai một lần nữa xác nhận Top 10 Khu vực có Giá nhà cao nhất dựa trên mã Zip. Những biểu đồ này là minh chứng trực quan cho thấy sự tập trung của những bất động sản có giá trị đầu tư lớn tại các khu vực cụ thể.

Biểu đồ thứ ba, so sánh Giá nhà (tuyệt đối) và Giá trên mỗi mét vuông (tương đối), là điểm mấu chốt mang đến chiều sâu phân tích. Biểu đồ này củng cố nhận định đã có: giá trị tổng thể của một căn nhà không hoàn toàn đồng nghĩa với giá trị trên mỗi đơn vị diện tích. Sự khác biệt về chiều cao cột giữa hai biểu đồ con cho cùng một mã Zip cho thấy rằng các yếu tố ngoài diện tích như vị trí đắc địa, chất lượng xây dựng hay tiện ích đi kèm đóng vai trò quan trọng trong việc định giá.

Đối với nhà đầu tư, sự tương phản này là thông tin quý giá. Nó nhắc nhở rằng việc đánh giá tiềm năng sinh lời không chỉ dừng lại ở giá bán tổng thể mà cần đi sâu vào phân tích hiệu quả sử dụng diện tích thông qua chỉ số giá trên mỗi mét vuông.

Tóm lại, truyền tải thông điệp quan trọng về sự phức tạp của việc định giá bất động sản, khuyến khích người xem vượt ra khỏi giá trị bề mặt để hiểu rõ hơn về cấu trúc giá trị thực.
