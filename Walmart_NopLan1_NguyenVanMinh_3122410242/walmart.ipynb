## 2. TI·ªÄN X·ª¨ L√ù D·ªÆ LI·ªÜU (DATA PREPROCESSING)

### 2.1. Ki·ªÉm tra Missing Values

```python
# Ki·ªÉm tra missing values trong t·ª´ng dataset
print("=" * 80)
print("MISSING VALUES ANALYSIS")
print("=" * 80)

print("\n1. TRAIN DATASET:")
print("-" * 80)
train_missing = train_df.isnull().sum()
train_missing_pct = (train_df.isnull().sum() / len(train_df)) * 100
train_missing_df = pd.DataFrame({
    'Missing_Count': train_missing,
    'Percentage': train_missing_pct
})
print(train_missing_df[train_missing_df['Missing_Count'] > 0])
if train_missing_df['Missing_Count'].sum() == 0:
    print("‚úì Kh√¥ng c√≥ missing values!")

print("\n2. FEATURES DATASET:")
print("-" * 80)
features_missing = features_df.isnull().sum()
features_missing_pct = (features_df.isnull().sum() / len(features_df)) * 100
features_missing_df = pd.DataFrame({
    'Missing_Count': features_missing,
    'Percentage': features_missing_pct
})
print(features_missing_df[features_missing_df['Missing_Count'] > 0])

print("\n3. STORES DATASET:")
print("-" * 80)
stores_missing = stores_df.isnull().sum()
stores_missing_pct = (stores_df.isnull().sum() / len(stores_df)) * 100
stores_missing_df = pd.DataFrame({
    'Missing_Count': stores_missing,
    'Percentage': stores_missing_pct
})
print(stores_missing_df[stores_missing_df['Missing_Count'] > 0])
if stores_missing_df['Missing_Count'].sum() == 0:
    print("‚úì Kh√¥ng c√≥ missing values!")
```

### 2.2. X·ª≠ l√Ω Missing Values

```python
# X·ª≠ l√Ω missing values trong features_df
# MarkDown columns c√≥ nhi·ªÅu missing values - fill b·∫±ng 0 (kh√¥ng c√≥ khuy·∫øn m√£i)
features_df['MarkDown1'].fillna(0, inplace=True)
features_df['MarkDown2'].fillna(0, inplace=True)
features_df['MarkDown3'].fillna(0, inplace=True)
features_df['MarkDown4'].fillna(0, inplace=True)
features_df['MarkDown5'].fillna(0, inplace=True)

# CPI v√† Unemployment: fill b·∫±ng gi√° tr·ªã trung b√¨nh theo Store
features_df['CPI'].fillna(features_df.groupby('Store')['CPI'].transform('mean'), inplace=True)
features_df['Unemployment'].fillna(features_df.groupby('Store')['Unemployment'].transform('mean'), inplace=True)

print("=" * 80)
print("X·ª¨ L√ù MISSING VALUES")
print("=" * 80)
print("‚úì ƒê√£ fill MarkDown1-5 b·∫±ng 0 (kh√¥ng c√≥ khuy·∫øn m√£i)")
print("‚úì ƒê√£ fill CPI v√† Unemployment b·∫±ng gi√° tr·ªã trung b√¨nh theo Store")
print("\nKi·ªÉm tra l·∫°i missing values:")
print(features_df.isnull().sum())
```

### 2.3. Chuy·ªÉn ƒë·ªïi ki·ªÉu d·ªØ li·ªáu v√† t·∫°o features m·ªõi

```python
# Chuy·ªÉn ƒë·ªïi Date sang datetime
train_df['Date'] = pd.to_datetime(train_df['Date'])
features_df['Date'] = pd.to_datetime(features_df['Date'])

# T·∫°o features t·ª´ Date cho train_df
train_df['Year'] = train_df['Date'].dt.year
train_df['Month'] = train_df['Date'].dt.month
train_df['Week'] = train_df['Date'].dt.isocalendar().week
train_df['Quarter'] = train_df['Date'].dt.quarter
train_df['DayOfYear'] = train_df['Date'].dt.dayofyear

# T·∫°o features t·ª´ Date cho features_df
features_df['Year'] = features_df['Date'].dt.year
features_df['Month'] = features_df['Date'].dt.month
features_df['Week'] = features_df['Date'].dt.isocalendar().week
features_df['Quarter'] = features_df['Date'].dt.quarter

# Chuy·ªÉn IsHoliday sang int (0/1)
train_df['IsHoliday'] = train_df['IsHoliday'].astype(int)
features_df['IsHoliday'] = features_df['IsHoliday'].astype(int)

print("=" * 80)
print("T·∫†O FEATURES M·ªöI")
print("=" * 80)
print("‚úì ƒê√£ chuy·ªÉn ƒë·ªïi Date sang datetime")
print("‚úì ƒê√£ t·∫°o features: Year, Month, Week, Quarter, DayOfYear")
print("‚úì ƒê√£ chuy·ªÉn IsHoliday sang int (0/1)")
print("\nC√°c features m·ªõi trong train_df:")
print(train_df[['Date', 'Year', 'Month', 'Week', 'Quarter', 'IsHoliday']].head())
```

### 2.4. Merge c√°c datasets

```python
# Merge train_df v·ªõi stores_df
df = train_df.merge(stores_df, on='Store', how='left')

# Merge v·ªõi features_df
df = df.merge(features_df, on=['Store', 'Date'], how='left', suffixes=('', '_feat'))

# X√≥a c·ªôt tr√πng l·∫∑p (IsHoliday_feat)
if 'IsHoliday_feat' in df.columns:
    df.drop('IsHoliday_feat', axis=1, inplace=True)

# Ki·ªÉm tra k·∫øt qu·∫£
print("=" * 80)
print("MERGE DATASETS")
print("=" * 80)
print(f"‚úì ƒê√£ merge th√†nh c√¥ng!")
print(f"Dataset ho√†n ch·ªânh: {df.shape[0]} rows √ó {df.shape[1]} columns")
print(f"\nC√°c c·ªôt trong dataset:")
print(df.columns.tolist())
print(f"\nM·∫´u d·ªØ li·ªáu sau khi merge:")
print(df.head())
```

```python
# Ki·ªÉm tra missing values sau khi merge
print("=" * 80)
print("KI·ªÇM TRA MISSING VALUES SAU KHI MERGE")
print("=" * 80)
missing_after_merge = df.isnull().sum()
missing_after_merge_pct = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({
    'Missing_Count': missing_after_merge,
    'Percentage': missing_after_merge_pct
})
print(missing_df[missing_df['Missing_Count'] > 0])

# Fill missing values c√≤n l·∫°i
df.fillna(df.median(numeric_only=True), inplace=True)
print("\n‚úì ƒê√£ fill t·∫•t c·∫£ missing values b·∫±ng median!")
print(f"Missing values c√≤n l·∫°i: {df.isnull().sum().sum()}")
```

## 5. PH·∫¶N 2 - PH√ÇN T√çCH SO S√ÅNH

### C√¢u h·ªèi 1: Doanh s·ªë trong tu·∫ßn l·ªÖ c√≥ cao h∆°n tu·∫ßn th∆∞·ªùng kh√¥ng? Ch√™nh l·ªách bao nhi√™u %?

```python
print("=" * 80)
print("C√ÇU H·ªéI 1: SO S√ÅNH DOANH S·ªê TU·∫¶N L·ªÑ vs TU·∫¶N TH∆Ø·ªúNG")
print("=" * 80)

holiday_stats = df.groupby('IsHoliday')['Weekly_Sales'].agg(['mean', 'median', 'sum', 'count', 'std'])
holiday_stats.index = ['Tu·∫ßn th∆∞·ªùng (0)', 'Tu·∫ßn l·ªÖ (1)']
print(holiday_stats)

# T√≠nh % ch√™nh l·ªách
mean_normal = df[df['IsHoliday']==0]['Weekly_Sales'].mean()
mean_holiday = df[df['IsHoliday']==1]['Weekly_Sales'].mean()
pct_diff = ((mean_holiday - mean_normal) / mean_normal) * 100

# T-test ƒë·ªÉ ki·ªÉm tra s·ª± kh√°c bi·ªát c√≥ √Ω nghƒ©a th·ªëng k√™
from scipy import stats
normal_sales = df[df['IsHoliday']==0]['Weekly_Sales']
holiday_sales = df[df['IsHoliday']==1]['Weekly_Sales']
t_stat, p_value = stats.ttest_ind(normal_sales, holiday_sales)

print("\nüìä PH√ÇN T√çCH:")
print(f"- Doanh s·ªë TB tu·∫ßn th∆∞·ªùng: ${mean_normal:,.2f}")
print(f"- Doanh s·ªë TB tu·∫ßn l·ªÖ: ${mean_holiday:,.2f}")
print(f"- Ch√™nh l·ªách: {pct_diff:+.2f}% ({'cao h∆°n' if pct_diff > 0 else 'th·∫•p h∆°n'})")
print(f"- T-statistic: {t_stat:.4f}, P-value: {p_value:.6f}")
print(f"- K·∫øt lu·∫≠n: S·ª± kh√°c bi·ªát {'C√ì' if p_value < 0.05 else 'KH√îNG C√ì'} √Ω nghƒ©a th·ªëng k√™ (Œ±=0.05)")
```

### C√¢u h·ªèi 2: Type A, B, C kh√°c bi·ªát nh∆∞ th·∫ø n√†o v·ªÅ doanh thu v√† quy m√¥?

```python
print("=" * 80)
print("C√ÇU H·ªéI 2: SO S√ÅNH CHI TI·∫æT THEO LO·∫†I C·ª¨A H√ÄNG")
print("=" * 80)

type_detail = df.groupby('Type').agg({
    'Weekly_Sales': ['mean', 'median', 'sum', 'std'],
    'Size': ['mean', 'min', 'max'],
    'Store': 'nunique'
}).round(2)

print(type_detail)

# ANOVA test ƒë·ªÉ ki·ªÉm tra s·ª± kh√°c bi·ªát gi·ªØa c√°c nh√≥m
type_a_sales = df[df['Type']=='A']['Weekly_Sales']
type_b_sales = df[df['Type']=='B']['Weekly_Sales']
type_c_sales = df[df['Type']=='C']['Weekly_Sales'] if 'C' in df['Type'].unique() else []

if len(type_c_sales) > 0:
    f_stat, p_value = stats.f_oneway(type_a_sales, type_b_sales, type_c_sales)
else:
    f_stat, p_value = stats.f_oneway(type_a_sales, type_b_sales)

print("\nüìä PH√ÇN T√çCH:")
print(f"- ANOVA F-statistic: {f_stat:.4f}, P-value: {p_value:.6f}")
print(f"- K·∫øt lu·∫≠n: S·ª± kh√°c bi·ªát gi·ªØa c√°c Type {'C√ì' if p_value < 0.05 else 'KH√îNG C√ì'} √Ω nghƒ©a th·ªëng k√™")

for store_type in df['Type'].unique():
    avg_sales = df[df['Type']==store_type]['Weekly_Sales'].mean()
    avg_size = df[df['Type']==store_type]['Size'].mean()
    num_stores = df[df['Type']==store_type]['Store'].nunique()
    print(f"- Type {store_type}: {num_stores} c·ª≠a h√†ng, Size TB: {avg_size:,.0f}, Doanh s·ªë TB: ${avg_sales:,.2f}")
```

### C√¢u h·ªèi 3: Qu√Ω n√†o trong nƒÉm c√≥ doanh s·ªë cao nh·∫•t? Th√°ng n√†o b√°n ch·∫°y nh·∫•t?

```python
print("=" * 80)
print("C√ÇU H·ªéI 3: PH√ÇN T√çCH THEO M√ôA V·ª§")
print("=" * 80)

# Theo Qu√Ω
quarterly_sales = df.groupby('Quarter')['Weekly_Sales'].agg(['mean', 'sum', 'count']).round(2)
quarterly_sales.index = ['Qu√Ω 1 (Q1)', 'Qu√Ω 2 (Q2)', 'Qu√Ω 3 (Q3)', 'Qu√Ω 4 (Q4)']
print("DOANH S·ªê THEO QU√ù:")
print(quarterly_sales)

best_quarter = df.groupby('Quarter')['Weekly_Sales'].mean().idxmax()
best_q_sales = df.groupby('Quarter')['Weekly_Sales'].mean().max()

# Theo Th√°ng
monthly_sales = df.groupby('Month')['Weekly_Sales'].agg(['mean', 'sum', 'count']).round(2)
month_names = ['Th√°ng 1', 'Th√°ng 2', 'Th√°ng 3', 'Th√°ng 4', 'Th√°ng 5', 'Th√°ng 6',
               'Th√°ng 7', 'Th√°ng 8', 'Th√°ng 9', 'Th√°ng 10', 'Th√°ng 11', 'Th√°ng 12']
monthly_sales.index = [month_names[i-1] for i in monthly_sales.index]
print("\nDOANH S·ªê THEO TH√ÅNG:")
print(monthly_sales)

best_month = df.groupby('Month')['Weekly_Sales'].mean().idxmax()
best_m_sales = df.groupby('Month')['Weekly_Sales'].mean().max()

print("\nüìä PH√ÇN T√çCH:")
print(f"- Qu√Ω c√≥ doanh s·ªë cao nh·∫•t: Qu√Ω {best_quarter} v·ªõi ${best_q_sales:,.2f} TB/tu·∫ßn")
print(f"- Th√°ng c√≥ doanh s·ªë cao nh·∫•t: Th√°ng {best_month} v·ªõi ${best_m_sales:,.2f} TB/tu·∫ßn")
print(f"- Xu h∆∞·ªõng: {'M√πa cu·ªëi nƒÉm' if best_quarter >= 3 else 'ƒê·∫ßu nƒÉm'} c√≥ doanh s·ªë t·ªët h∆°n")
```

### C√¢u h·ªèi 4: C·ª≠a h√†ng l·ªõn h∆°n c√≥ doanh thu cao h∆°n kh√¥ng?

```python
print("=" * 80)
print("C√ÇU H·ªéI 4: T∆Ø∆†NG QUAN GI·ªÆA QUY M√î C·ª¨A H√ÄNG V√Ä DOANH S·ªê")
print("=" * 80)

# T√≠nh correlation
store_size_sales = df.groupby('Store').agg({
    'Size': 'first',
    'Weekly_Sales': 'mean'
})

correlation = store_size_sales['Size'].corr(store_size_sales['Weekly_Sales'])
print(f"H·ªá s·ªë t∆∞∆°ng quan Pearson (Size vs Weekly_Sales): {correlation:.4f}")

# Chia c·ª≠a h√†ng th√†nh nh√≥m theo quy m√¥
size_quartiles = store_size_sales['Size'].quantile([0.25, 0.5, 0.75])
print(f"\nPh√¢n v·ªã quy m√¥ c·ª≠a h√†ng:")
print(f"- Q1 (25%): {size_quartiles[0.25]:,.0f} sqft")
print(f"- Q2 (50% - Median): {size_quartiles[0.5]:,.0f} sqft")
print(f"- Q3 (75%): {size_quartiles[0.75]:,.0f} sqft")

# Ph√¢n lo·∫°i v√† so s√°nh
df['Size_Category'] = pd.cut(df['Size'], 
                               bins=[0, size_quartiles[0.25], size_quartiles[0.5], size_quartiles[0.75], float('inf')],
                               labels=['Nh·ªè', 'Trung b√¨nh', 'L·ªõn', 'R·∫•t l·ªõn'])

size_category_sales = df.groupby('Size_Category')['Weekly_Sales'].agg(['mean', 'median', 'count'])
print("\nDOANH S·ªê THEO NH√ìM QUY M√î:")
print(size_category_sales)

print("\nüìä PH√ÇN T√çCH:")
if correlation > 0.5:
    strength = "m·∫°nh"
elif correlation > 0.3:
    strength = "trung b√¨nh"
else:
    strength = "y·∫øu"
print(f"- T∆∞∆°ng quan {strength} v√† {'d∆∞∆°ng' if correlation > 0 else '√¢m'} (r={correlation:.4f})")
print(f"- C·ª≠a h√†ng l·ªõn h∆°n {'C√ì XU H∆Ø·ªöNG' if correlation > 0.3 else 'KH√îNG R√ï R√ÄNG'} c√≥ doanh s·ªë cao h∆°n")
```

## 8. PH·∫¶N 4 - PH√ÇN C·ª§M (CLUSTERING ANALYSIS)

### 8.1. Chu·∫©n b·ªã d·ªØ li·ªáu cho Clustering

```python
# T·∫°o dataset aggregated cho clustering (theo Store)
store_features = df.groupby('Store').agg({
    'Weekly_Sales': ['mean', 'std', 'sum'],
    'Size': 'first',
    'Temperature': 'mean',
    'Fuel_Price': 'mean',
    'CPI': 'mean',
    'Unemployment': 'mean',
    'IsHoliday': 'sum',
    'Type': 'first'
}).reset_index()

# Flatten column names
store_features.columns = ['Store', 'Avg_Sales', 'Std_Sales', 'Total_Sales', 'Size', 
                          'Avg_Temp', 'Avg_Fuel', 'Avg_CPI', 'Avg_Unemp', 'Holiday_Count', 'Type']

# Ch·ªçn features cho clustering
clustering_features = ['Avg_Sales', 'Std_Sales', 'Size', 'Avg_Temp', 'Avg_Fuel', 'Avg_CPI', 'Avg_Unemp']
X_cluster = store_features[clustering_features].copy()

# Chu·∫©n h√≥a d·ªØ li·ªáu
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

print("=" * 80)
print("CHU·∫®N B·ªä D·ªÆ LI·ªÜU CHO CLUSTERING")
print("=" * 80)
print(f"S·ªë c·ª≠a h√†ng: {len(store_features)}")
print(f"S·ªë features: {len(clustering_features)}")
print(f"\nFeatures s·ª≠ d·ª•ng:")
for i, feat in enumerate(clustering_features, 1):
    print(f"  {i}. {feat}")
print(f"\nD·ªØ li·ªáu sau chu·∫©n h√≥a: {X_scaled.shape}")
print("\n‚úì D·ªØ li·ªáu ƒë√£ s·∫µn s√†ng cho ph√¢n c·ª•m!")
```

### 8.2. K-Means Clustering

```python
# Elbow method ƒë·ªÉ t√¨m s·ªë cluster t·ªëi ∆∞u
inertias = []
silhouette_scores = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# V·∫Ω Elbow curve
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

ax1.plot(K_range, inertias, 'bo-', linewidth=2, markersize=8)
ax1.set_xlabel('S·ªë l∆∞·ª£ng Clusters (k)', fontsize=12)
ax1.set_ylabel('Inertia', fontsize=12)
ax1.set_title('Elbow Method - T√¨m s·ªë cluster t·ªëi ∆∞u', fontweight='bold', fontsize=14)
ax1.grid(True, alpha=0.3)

ax2.plot(K_range, silhouette_scores, 'ro-', linewidth=2, markersize=8)
ax2.set_xlabel('S·ªë l∆∞·ª£ng Clusters (k)', fontsize=12)
ax2.set_ylabel('Silhouette Score', fontsize=12)
ax2.set_title('Silhouette Score theo s·ªë Clusters', fontweight='bold', fontsize=14)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Ch·ªçn k=3 d·ª±a tr√™n elbow method
optimal_k = 3
print(f"\nüìä Ch·ªçn s·ªë cluster t·ªëi ∆∞u: k = {optimal_k}")

# √Åp d·ª•ng K-Means v·ªõi k=3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
store_features['Cluster_KMeans'] = kmeans.fit_predict(X_scaled)

print("\n" + "=" * 80)
print(f"K-MEANS CLUSTERING (k={optimal_k})")
print("=" * 80)

for cluster in range(optimal_k):
    cluster_data = store_features[store_features['Cluster_KMeans'] == cluster]
    print(f"\nCLUSTER {cluster}:")
    print(f"  - S·ªë c·ª≠a h√†ng: {len(cluster_data)}")
    print(f"  - Doanh s·ªë TB: ${cluster_data['Avg_Sales'].mean():,.2f}")
    print(f"  - Size TB: {cluster_data['Size'].mean():,.0f} sqft")
    print(f"  - Stores: {cluster_data['Store'].tolist()}")

silhouette_avg = silhouette_score(X_scaled, store_features['Cluster_KMeans'])
print(f"\nSilhouette Score: {silhouette_avg:.4f}")
print("‚úì K-Means clustering ho√†n t·∫•t!")
```

### 8.3. Hierarchical Clustering

```python
# Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=3, linkage='ward')
store_features['Cluster_Hierarchical'] = hierarchical.fit_predict(X_scaled)

# V·∫Ω Dendrogram
plt.figure(figsize=(16, 8))
linkage_matrix = linkage(X_scaled, method='ward')
dendrogram(linkage_matrix, labels=store_features['Store'].values, leaf_font_size=10)
plt.title('DENDROGRAM - HIERARCHICAL CLUSTERING', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Store ID', fontsize=12)
plt.ylabel('Euclidean Distance', fontsize=12)
plt.axhline(y=15, color='r', linestyle='--', label='Cut threshold')
plt.legend()
plt.tight_layout()
plt.show()

print("=" * 80)
print("HIERARCHICAL CLUSTERING (n_clusters=3)")
print("=" * 80)

for cluster in range(3):
    cluster_data = store_features[store_features['Cluster_Hierarchical'] == cluster]
    print(f"\nCLUSTER {cluster}:")
    print(f"  - S·ªë c·ª≠a h√†ng: {len(cluster_data)}")
    print(f"  - Doanh s·ªë TB: ${cluster_data['Avg_Sales'].mean():,.2f}")
    print(f"  - Stores: {sorted(cluster_data['Store'].tolist())}")

h_silhouette = silhouette_score(X_scaled, store_features['Cluster_Hierarchical'])
print(f"\nSilhouette Score: {h_silhouette:.4f}")
print("‚úì Hierarchical clustering ho√†n t·∫•t!")
```

### 8.4. DBSCAN Clustering

```python
# DBSCAN
dbscan = DBSCAN(eps=2.5, min_samples=3)
store_features['Cluster_DBSCAN'] = dbscan.fit_predict(X_scaled)

print("=" * 80)
print("DBSCAN CLUSTERING (eps=2.5, min_samples=3)")
print("=" * 80)

unique_clusters = sorted(store_features['Cluster_DBSCAN'].unique())
print(f"\nS·ªë clusters t√¨m ƒë∆∞·ª£c: {len([c for c in unique_clusters if c != -1])}")
print(f"S·ªë noise points: {len(store_features[store_features['Cluster_DBSCAN'] == -1])}")

for cluster in unique_clusters:
    if cluster == -1:
        cluster_name = "NOISE"
    else:
        cluster_name = f"CLUSTER {cluster}"
    
    cluster_data = store_features[store_features['Cluster_DBSCAN'] == cluster]
    print(f"\n{cluster_name}:")
    print(f"  - S·ªë c·ª≠a h√†ng: {len(cluster_data)}")
    if cluster != -1:
        print(f"  - Doanh s·ªë TB: ${cluster_data['Avg_Sales'].mean():,.2f}")
    print(f"  - Stores: {sorted(cluster_data['Store'].tolist())}")

# Silhouette score (ch·ªâ t√≠nh cho non-noise points)
if len(unique_clusters) > 1 and len(store_features[store_features['Cluster_DBSCAN'] != -1]) > 0:
    non_noise_mask = store_features['Cluster_DBSCAN'] != -1
    if len(store_features[non_noise_mask]['Cluster_DBSCAN'].unique()) > 1:
        dbscan_silhouette = silhouette_score(X_scaled[non_noise_mask], 
                                              store_features[non_noise_mask]['Cluster_DBSCAN'])
        print(f"\nSilhouette Score (non-noise): {dbscan_silhouette:.4f}")

print("\n‚úì DBSCAN clustering ho√†n t·∫•t!")
```

### 8.5. Tr·ª±c quan h√≥a k·∫øt qu·∫£ Clustering

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('SO S√ÅNH C√ÅC PH∆Ø∆†NG PH√ÅP CLUSTERING', fontsize=16, fontweight='bold')

# K-Means
axes[0].scatter(store_features['Avg_Sales'], store_features['Size'], 
                c=store_features['Cluster_KMeans'], cmap='viridis', s=150, alpha=0.7, edgecolors='black')
axes[0].set_xlabel('Average Sales ($)', fontsize=11)
axes[0].set_ylabel('Store Size (sqft)', fontsize=11)
axes[0].set_title(f'K-Means (Silhouette: {silhouette_avg:.3f})', fontweight='bold')
for i, txt in enumerate(store_features['Store']):
    axes[0].annotate(txt, (store_features['Avg_Sales'].iloc[i], store_features['Size'].iloc[i]), 
                     fontsize=8, ha='center')

# Hierarchical
axes[1].scatter(store_features['Avg_Sales'], store_features['Size'], 
                c=store_features['Cluster_Hierarchical'], cmap='plasma', s=150, alpha=0.7, edgecolors='black')
axes[1].set_xlabel('Average Sales ($)', fontsize=11)
axes[1].set_ylabel('Store Size (sqft)', fontsize=11)
axes[1].set_title(f'Hierarchical (Silhouette: {h_silhouette:.3f})', fontweight='bold')
for i, txt in enumerate(store_features['Store']):
    axes[1].annotate(txt, (store_features['Avg_Sales'].iloc[i], store_features['Size'].iloc[i]), 
                     fontsize=8, ha='center')

# DBSCAN
scatter = axes[2].scatter(store_features['Avg_Sales'], store_features['Size'], 
                          c=store_features['Cluster_DBSCAN'], cmap='coolwarm', s=150, alpha=0.7, edgecolors='black')
axes[2].set_xlabel('Average Sales ($)', fontsize=11)
axes[2].set_ylabel('Store Size (sqft)', fontsize=11)
axes[2].set_title('DBSCAN', fontweight='bold')
for i, txt in enumerate(store_features['Store']):
    axes[2].annotate(txt, (store_features['Avg_Sales'].iloc[i], store_features['Size'].iloc[i]), 
                     fontsize=8, ha='center')

plt.tight_layout()
plt.show()

print("‚úì ƒê√£ t·∫°o bi·ªÉu ƒë·ªì so s√°nh clustering")
```

## 9. PH·∫¶N 5 - PH√ÇN LO·∫†I (CLASSIFICATION ANALYSIS)

### 9.1. T·∫°o bi·∫øn Target v√† chu·∫©n b·ªã d·ªØ li·ªáu

```python
# T·∫°o target variable: Ph√¢n lo·∫°i doanh s·ªë cao/th·∫•p
# S·ª≠ d·ª•ng median l√†m ng∆∞·ª°ng
median_sales = df['Weekly_Sales'].median()
df['Sales_Category'] = (df['Weekly_Sales'] > median_sales).astype(int)  # 1: High, 0: Low

# Ch·ªçn features cho classification
classification_features = ['Store', 'Dept', 'Size', 'Temperature', 'Fuel_Price', 
                           'CPI', 'Unemployment', 'IsHoliday', 'Month', 'Quarter']

# Encode Type
le_type = LabelEncoder()
df['Type_Encoded'] = le_type.fit_transform(df['Type'])
classification_features.append('Type_Encoded')

# T·∫°o X v√† y
X = df[classification_features].copy()
y = df['Sales_Category'].copy()

# Chia train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print("=" * 80)
print("CHU·∫®N B·ªä D·ªÆ LI·ªÜU CHO CLASSIFICATION")
print("=" * 80)
print(f"Ng∆∞·ª°ng ph√¢n lo·∫°i (median): ${median_sales:,.2f}")
print(f"\nPh√¢n ph·ªëi Target Variable:")
print(f"  - Low Sales (0): {(y == 0).sum():,} ({(y == 0).sum()/len(y)*100:.1f}%)")
print(f"  - High Sales (1): {(y == 1).sum():,} ({(y == 1).sum()/len(y)*100:.1f}%)")
print(f"\nK√≠ch th∆∞·ªõc dataset:")
print(f"  - Training set: {len(X_train):,} samples")
print(f"  - Test set: {len(X_test):,} samples")
print(f"  - S·ªë features: {len(classification_features)}")
print("\n‚úì D·ªØ li·ªáu ƒë√£ s·∫µn s√†ng cho ph√¢n lo·∫°i!")
```

### 9.2. Decision Tree Classifier

```python
# Train Decision Tree
dt_classifier = DecisionTreeClassifier(max_depth=5, random_state=42, min_samples_split=100)
dt_classifier.fit(X_train, y_train)

# Predictions
y_pred_dt = dt_classifier.predict(X_test)

# Evaluation
dt_accuracy = dt_classifier.score(X_test, y_test)

print("=" * 80)
print("DECISION TREE CLASSIFIER")
print("=" * 80)
print(f"\nAccuracy: {dt_accuracy:.4f} ({dt_accuracy*100:.2f}%)")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_dt, target_names=['Low Sales', 'High Sales']))

print("\nConfusion Matrix:")
cm_dt = confusion_matrix(y_test, y_pred_dt)
print(cm_dt)

# Feature Importance
feature_importance_dt = pd.DataFrame({
    'Feature': classification_features,
    'Importance': dt_classifier.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nTop 10 Features quan tr·ªçng nh·∫•t:")
print(feature_importance_dt.head(10))

print("\n‚úì Decision Tree ho√†n t·∫•t!")
```

### 9.3. Random Forest Classifier

```python
# Train Random Forest
rf_classifier = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)
rf_classifier.fit(X_train, y_train)

# Predictions
y_pred_rf = rf_classifier.predict(X_test)

# Evaluation
rf_accuracy = rf_classifier.score(X_test, y_test)

print("=" * 80)
print("RANDOM FOREST CLASSIFIER")
print("=" * 80)
print(f"\nAccuracy: {rf_accuracy:.4f} ({rf_accuracy*100:.2f}%)")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf, target_names=['Low Sales', 'High Sales']))

print("\nConfusion Matrix:")
cm_rf = confusion_matrix(y_test, y_pred_rf)
print(cm_rf)

# Feature Importance
feature_importance_rf = pd.DataFrame({
    'Feature': classification_features,
    'Importance': rf_classifier.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nTop 10 Features quan tr·ªçng nh·∫•t:")
print(feature_importance_rf.head(10))

print("\n‚úì Random Forest ho√†n t·∫•t!")
```

### 9.4. Logistic Regression

```python
# Chu·∫©n h√≥a d·ªØ li·ªáu cho Logistic Regression
scaler_lr = StandardScaler()
X_train_scaled = scaler_lr.fit_transform(X_train)
X_test_scaled = scaler_lr.transform(X_test)

# Train Logistic Regression
lr_classifier = LogisticRegression(max_iter=1000, random_state=42)
lr_classifier.fit(X_train_scaled, y_train)

# Predictions
y_pred_lr = lr_classifier.predict(X_test_scaled)

# Evaluation
lr_accuracy = lr_classifier.score(X_test_scaled, y_test)

print("=" * 80)
print("LOGISTIC REGRESSION")
print("=" * 80)
print(f"\nAccuracy: {lr_accuracy:.4f} ({lr_accuracy*100:.2f}%)")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_lr, target_names=['Low Sales', 'High Sales']))

print("\nConfusion Matrix:")
cm_lr = confusion_matrix(y_test, y_pred_lr)
print(cm_lr)

print("\n‚úì Logistic Regression ho√†n t·∫•t!")
```
