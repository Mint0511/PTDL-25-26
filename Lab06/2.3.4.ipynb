### 1. Import thư viện và nạp dữ liệu vào notebook

Vì bộ dữ liệu mushrooms.csv chứa hoàn toàn dữ liệu dạng phân loại (categorical) – ví dụ: 'x', 's', 'n', chúng ta sẽ sử dụng mô hình CategoricalNB (Naive Bayes cho dữ liệu phân loại) thay vì GaussianNB.

```python
# Import các thư viện cần thiết
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder # Cần thiết để mã hóa
from sklearn.naive_bayes import CategoricalNB # Thay vì GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Nạp dữ liệu
data = pd.read_csv('data/mushrooms.csv')
```

```python
# Hiển thị 5 dòng đầu tiên
print(data.head())
```

bộ dữ liệu gồm 5 cột và 400 dòng, trong đó:

* **User ID**: Mã định danh duy nhất cho mỗi khách hàng
* **Gender**: Giới tính của khách hàng (Male hoặc Female)
* **Age**: Tuổi của khách hàng (giá trị số nguyên từ 18 đến 60)
* **EstimatedSalary**: Mức lương ước tính của khách hàng (giá trị số từ 15,000 đến 150,000)
* **Purchased**: Nhãn mục tiêu: 0 nếu khách hàng không mua sản phẩm, 1 nếu đã mua

### 2. Xử lý dữ liệu trước khi xây dựng mô hình từ dữ liệu

Dữ liệu này cần nhiều bước xử lý hơn:

1. Phát hiện và loại bỏ các giá trị bị thiếu (được biểu thị bằng dấu '?').

2. Tách features (X) và target (y). Target ở đây là cột class.

3. Mã hóa toàn bộ dữ liệu (cả X và y) từ dạng chữ (object) sang dạng số mà mô hình CategoricalNB có thể hiểu được.

4. Chia tập dữ liệu (80% train, 20% test).

```python
# 1. Xử lý giá trị thiếu ('?')
# Phát hiện cột 'stalk-root' có 2480 giá trị '?'
print((data == '?').sum()) 
# Thay thế '?' bằng NaN và xóa các hàng đó
data.replace('?', np.nan, inplace=True)
data.dropna(inplace=True)
# Kích thước dữ liệu còn lại: (5644, 23)
```

```python
# 2. Tách features (X) và target (y)
X = data.drop('class', axis=1)
y = data['class']
```

```python
# 3. Mã hóa dữ liệu
# Mã hóa X (tất cả các cột features) bằng OrdinalEncoder
encoder_X = OrdinalEncoder()
X_encoded = encoder_X.fit_transform(X)

# Mã hóa y (target) bằng LabelEncoder (e -> 0, p -> 1)
encoder_y = LabelEncoder()
y_encoded = encoder_y.fit_transform(y)
```

```python
# 4. Chia dữ liệu
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)
```

```python
data
```

### 3. Xây dựng mô hình Naïve Bayes

Chúng ta sử dụng CategoricalNB vì tất cả các features đầu vào đều là dạng phân loại đã được mã hóa thành các số nguyên.

```python
# Khởi tạo mô hình Categorical Naïve Bayes
classifier = CategoricalNB()
```

```python
# Huấn luyện mô hình trên tập X_train
classifier.fit(X_train, y_train)
```

### 4. Đánh giá hiệu quả của mô hình

Bước đánh giá này tương tự như ví dụ, sử dụng các hàm accuracy_score, confusion_matrix, và classification_report.

```python
# Đưa ra dự đoán trên tập test
y_pred = classifier.predict(X_test)
```

```python
# Lấy tên lớp gốc ('e', 'p') để hiển thị trong báo cáo
target_names = encoder_y.classes_
```

```python
# Đánh giá mô hình
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred, target_names=target_names)
```

```python
# In kết quả
print(f"Accuracy: {accuracy:.4f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(classification_rep)
```
